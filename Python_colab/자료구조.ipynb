## Queue 예제


```python
# List 사용
class ListQueue(object):
    def __init__(self):
        self.queue = []

    def enqueue(self,n):
        self.queue.append(n)

    def dequeue(self):
        if len(self.queue) == 0:
            return -1
        return self.queue.pop(0)

    def printQueue(self):
        print(self.queue)
```

## 해시 테이블

```python
class HashTable:
    def __init__(self,length=5):
        self.max_len = length
        self.table =[[] for _ in range(self.max_len)]

    def _hash(self,key):
        res = sum([ord(s) for s in key])
        return res % self.max_len

    def set(self,key,value):
        index = self._hash(key)
        self.table[index].append((key,value))

    def get(self,key):
        index = self._hash(key)
        value = self.table[index]
        if not value:
            return None
        for v in value:
            if v[0] ==key:
                return v[1]
        return None

if __name__ =="__main__":
    capital = HashTable()
    country = ["Korea","America","China","England","Turkieye"]
    city = ["Seoul","Wahington","Beijing","London","Ankara"]
    for co, ci in zip(country, city):
        capital.set(co,ci)
```

## 이진 트리 예제

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

    @staticmethod
    def init_tree():
        global root
        # 루트 노드 생성
        root = Node("A")
        # 2층
        root.left  = Node("B")
        root.right = Node("C")
        # 3층
        root.left.left  = Node("D")
        root.left.right = Node("E")
        root.right.left  = Node("F")
        root.right.right = Node("G")

    @staticmethod
    def preorder_traverse(node):
        if node is None:
            return
        print(node.data, end='->')
        Node.preorder_traverse(node.left)
        Node.preorder_traverse(node.right)

    @staticmethod
    def inorder_traverse(node):
        if node is None:
            return
        Node.inorder_traverse(node.left)
        print(node.data, end='->')
        Node.inorder_traverse(node.right)

    @staticmethod
    def postorder_traverse(node):
        if node is None:
            return
        Node.postorder_traverse(node.left)
        Node.postorder_traverse(node.right)
        print(node.data, end='->')


if __name__ == "__main__":
    # 트리 초기화
    Node.init_tree()
    # 전위 순회
    Node.preorder_traverse(root)
    print("\n\n------------\n")
    # 중위 순회
    Node.inorder_traverse(root)
    print("\n\n------------\n")
    # 후위 순회
    Node.postorder_traverse(root)
```
