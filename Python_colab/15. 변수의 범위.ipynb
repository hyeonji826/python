# 1. 범위(scope)
파이썬에서 변수의 범위(scope)는 해당 변수가 프로그램 내에서 참조되고 변경될 수 있는 영역을 의미합니다. 파이썬의 변수 범위는 크게 네 가지로 분류됩니다

### 1. 지역(Local) 범위
 가장 내부의 범위로, 함수 내에서 정의된 변수들이 해당됩니다. 함수 내에서만 접근할 수 있습니다.

```r
def local_example():
    local_var = "로컬 변수"
    print(local_var)

local_example()
```

````python
def local_example():
    local_var = "로컬 변수"
    print(local_var)

local_example()

# 밖에서 프린트 해보기
# print(local_var)      # 에러남 찌글찌글~~~
````

### 2. 함수 내부의 블록(Enclosing) 범위
 중첩된 함수 내부에서, 바깥쪽 함수의 변수들을 의미합니다. 중첩 함수가 바깥 함수의 변수를 사용할 수 있게 해줍니다.

```r
def outer_function():
    enclosing_var = "둘러싼 범위 변수"
    
    def inner_function():
        print(enclosing_var)
    
    inner_function()

outer_function()
```

````python
def outer_function():
    enclosing_var = "둘러싼 범위 변수"

    # 함수안에 함수 --> 자주 쓰임
    def inner_function():
        print(enclosing_var)

    inner_function()

# 메모리 구조 : 외부함수가 내부함수의 주소를 가리킴
outer_function()
````

### 3. 전역(Global) 범위
 스크립트 전체에서 사용되는 범위로, 함수 외부에 정의된 변수가 해당됩니다. 모든 함수에서 접근할 수 있지만, 함수 내부에서 수정하려면 global 키워드가 필요합니다.

```r
global_var = "I'm a global variable"

def test_global_scope():
    print(global_var)

test_global_scope()
```

````python
global_var = "I'm a global variable"

# 무조건 바깥에서 선언되면 def 안에다가 호출할 수 있음
def test_global_scope():
    print(global_var)

test_global_scope()
````

````python
#  함수 내부에서 수정하려면 global 키워드가 필요합니다. 의 예제 이해안되면 보기

global_var = 10

def modify_global():
    # 바깥 선언 값을 바꾸고 싶을 때, global을 사용하면 해당 함수가 바깥 함수로 취급함
    global global_var
    # 이름이 동일하대도 함수 내부에 있는게 우선시 되나, 바깥에 있는걸 없애진 않음
    global_var = 20

modify_global()
print(global_var)  # 출력: 20
````

### 4. 내장 범위(Built-in)
 파이썬이 기본적으로 제공하는 내장 함수나 예외들이 있는 범위로, 예를 들어 print(), len() 등이 해당됩니다.

```r
len = 5  # 내장 함수 'len'을 덮어씀
print(len([1, 2, 3]))  # 오류 발생! len이 함수가 아니라 변수로 인식됨
```

````python
len = 5  # 내장 함수 'len'을 덮어씀
print(len([1, 2, 3]))  # 오류 발생! len이 함수가 아니라 변수로 인식됨
````

# 2. 변수의 범위 탐색 순서

1. Local scope : 내부가 우선

2. Enclosing scope : 내부의 바깥에 감싸고 있는

3. Global scope : 전역변수

4. Built-in scope : 미리 만들어진거

> 따라서 지역 범위에 동일한 이름의 변수가 없으면 파이썬은 둘러싼 범위를 확인하고, 그다음으로 전역 범위, 마지막으로 내장 범위를 확인합니다.
- 변수의 범위를 이해하고 관리하는 것은 코드의 가독성과 유지 보수성을 높이고, 예기치 않은 오류를 방지하기 위함입니다.
