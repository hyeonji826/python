# 5. ì‚¬ìš©ì ì •ì˜ ì˜ˆì™¸ í´ë˜ìŠ¤ë¥¼ ì§ì ‘ ë§Œë“¤ê³  í™œìš©í•˜ê¸°

````python
class AgeLimitError(Exception):
    def __init__(self, age, message='MZì„¸ëŒ€ ë‚˜ì´ì˜ ë²”ìœ„ê°€ ì•„ë‹˜'):
        self.age = age
        self.message = message
        # ë¶€ëª¨ í´ë˜ìŠ¤ì—ê²Œ Exception('ë©”ì„¸ì§€~')ì™€ ë™ì¼í•œ ê¸°ëŠ¥
        super().__init__(self.message) # Exception('ë©”ì„¸ì§€')

def check_age(age):
    if age < 14:
        raise AgeLimitError(age, 'MZ ë‚˜ì´ ë²”ìœ„ë³´ë‹¤ ì‘ìŒ')
    elif age > 43:
        raise AgeLimitError(age, 'MZ ë‚˜ì´ ë²”ìœ„ë³´ë‹¤ í¼')
    else:
        return f'{age} ë‚˜ì´: MZ ë‚˜ì´ ë²”ìœ„ì•ˆì— í¬í•¨'
````

````python
ages = [17, 60, 46, 20, 80, 55, 26, 12, 5]

for age in ages:
    try:
        # í•¨ìˆ˜ì—ì„œ return í•´ì£¼ê¸°ë•Œë¬¸ì— printë¡œ ê²°ê³¼ê°’ ì¶œë ¥
        print(check_age(age))
    except AgeLimitError as e:
        print(f'{e.age} ë‚˜ì´ ì—ëŸ¬: {e}')
````
