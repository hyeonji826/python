### 3. \_\_len__()

íŒŒì´ì¬ì—ì„œ ê°ì²´ì˜ ê¸¸ì´ ë˜ëŠ” í¬ê¸°ë¥¼ ë°˜í™˜í•˜ëŠ” íŠ¹ë³„í•œ ë©”ì„œë“œë¡œ, len() í•¨ìˆ˜ê°€ í˜¸ì¶œë  ë•Œ ìë™ìœ¼ë¡œ í˜¸ì¶œë©ë‹ˆë‹¤.  ì´ ë©”ì„œë“œëŠ” ê°ì²´ì˜ í•­ëª© ìˆ˜ë¥¼ ì¸¡ì •í•˜ê±°ë‚˜, íŠ¹ì •í•œ í¬ê¸°(ì˜ˆ: ë¦¬ìŠ¤íŠ¸, ë¬¸ìì—´, íŠœí”Œ ë“±)ë¥¼ ë‚˜íƒ€ë‚´ê³ ì í•  ë•Œ êµ¬í˜„ë©ë‹ˆë‹¤.

### 4. \_\_getitem__()

íŒŒì´ì¬ì—ì„œ ì¸ë±ì‹±ì„ ì§€ì›í•˜ê¸° ìœ„í•´ ì‚¬ìš©ë˜ëŠ” íŠ¹ë³„í•œ ë©”ì„œë“œë¡œ, ê°ì²´ì˜ íŠ¹ì • í•­ëª©ì„ ê°€ì ¸ì˜¤ê¸° ìœ„í•´ ëŒ€ê´„í˜¸([])ë¥¼ ì‚¬ìš©í•  ë•Œ í˜¸ì¶œë©ë‹ˆë‹¤. ì´ ë©”ì„œë“œëŠ” ê°ì²´ì˜ íŠ¹ì • ì¸ë±ìŠ¤ë‚˜ í‚¤ì— í•´ë‹¹í•˜ëŠ” ê°’ì„ ë°˜í™˜í•˜ëŠ” ì—­í• ì„ í•©ë‹ˆë‹¤.

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # ë°˜ë“œì‹œ ì•Œê³ ìˆì!ğŸ¤©
    def __getitem__(self, index):
        if index == 0:
            return self.x
        elif index == 1:
            return self.y
        else:
            return -1

pt = Point(5, 3)
print(pt)
print(pt[0])
print(pt[1])
print(pt[100])
```

```python
class MyList:
    def __init__(self, data):
        self.data = data
    # ì¸ë±ìŠ¤ë¡œ ë½‘ì•„ì˜¬ ìˆ˜ ìˆê²Œí•´ì¤Œ
    def __getitem__(self, index):
        return self.data[index]
```

```python
ml = MyList([10, 20, 30, 40])
print(ml[0])
print(ml[1])
print(ml[-1])
```

### 5. \_\_call__()

íŒŒì´ì¬ì—ì„œ ê°ì²´ë¥¼ í•¨ìˆ˜ì²˜ëŸ¼ í˜¸ì¶œí•  ìˆ˜ ìˆê²Œ í•´ì£¼ëŠ” íŠ¹ë³„í•œ ë©”ì„œë“œì…ë‹ˆë‹¤. ì´ ë©”ì„œë“œë¥¼ êµ¬í˜„í•˜ë©´, í•´ë‹¹ í´ë˜ìŠ¤ì˜ ì¸ìŠ¤í„´ìŠ¤ë¥¼ í•¨ìˆ˜ì²˜ëŸ¼ ì‚¬ìš©í•  ìˆ˜ ìˆìœ¼ë©°, ì¸ìŠ¤í„´ìŠ¤ì— ëŒ€í•´ ê´„í˜¸(())ë¥¼ ì‚¬ìš©í•˜ì—¬ ê°’ì„ ì „ë‹¬í•˜ê³  ê²°ê³¼ë¥¼ ë°˜í™˜ë°›ì„ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

```python
class CallableObject:
    # kwargs : ë”•ì…”ë„ˆë¦¬ í˜•íƒœë¡œ ë°›ì„ ìˆ˜ ìˆê²Œí•¨ (í‚¤:ê°’)
    def __call__(self, *args, **kwargs):
        print(f'args:{args}, kwargs:{kwargs}')

# ê°ì²´ ìƒì„±
callable_obj = CallableObject()
# í•¨ìˆ˜ì²˜ëŸ¼ ì‚¬ìš©ê°€ëŠ¥í•¨
callable_obj(1, 2, 3, a='A', b='B')
```

### 1. ìº¡ìŠí™” (Encapsulation)
ì¤‘ìš”í•œ ë°ì´í„°ëŠ” ìˆ¨ê¸°ê³ , ì™¸ë¶€ì—ëŠ” ì¸í„°í˜ì´ìŠ¤ë§Œ ì—´ì–´ë‘¡ë‹ˆë‹¤.

```python
class Animal:
    # ìƒì„±ìì—ì„œ ì¸ìŠ¤í„´ìŠ¤ ë³€ìˆ˜ ìƒì„±
    def __init__(self, name, age):
        self.name = name
        self.__age = age  # __ë¡œ ì‹œì‘í•˜ë©´ ê°ì¶°ì§€ëŠ” private ì†ì„±

    def get_age(self):
        return self.__age

    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            print("ë‚˜ì´ëŠ” ì–‘ìˆ˜ì—¬ì•¼ í•©ë‹ˆë‹¤.")

class Dog(Animal):
    def __init__(self, name, age, breed):
        super().__init__(name, age)
        self.breed = breed

    def bark(self):
        print(f"{self.name}ê°€ ë©ë© ì§–ìŠµë‹ˆë‹¤!")

# ì‚¬ìš© ì˜ˆ
dog = Dog("ë£¨ì‹œ", 15, "í¬ë©”")
dog.bark()  # ë£¨ì‹œê°€ ë©ë© ì§–ìŠµë‹ˆë‹¤!

print(dog.name)      # ë£¨ì‹œ
print(dog.breed)     # í¬ë©”
print(dog.get_age())

dog.set_age(15)
print(dog.get_age()) # 15

dog.__age = 100      # ì™¸ë¶€ì—ì„œ ìˆ˜ì • ì‹œë„ (ì‹¤ì œ ì†ì„±ì€ ë°”ë€Œì§€ ì•ŠìŒ)
print(dog.get_age()) # ì—¬ì „íˆ 15
```

```python
class Animal:
    def speak(self):
        print("ë™ë¬¼ì´ ì†Œë¦¬ë¥¼ ëƒ…ë‹ˆë‹¤.")

class Dog(Animal):
    def speak(self):
        print("ë©ë©!")

# ì‚¬ìš© ì˜ˆ
a = Animal()
a.speak()  # ë™ë¬¼ì´ ì†Œë¦¬ë¥¼ ëƒ…ë‹ˆë‹¤.

rucy = Dog()
rucy.speak()  # ë©ë©! (ë¶€ëª¨ ë©”ì„œë“œë¥¼ ì˜¤ë²„ë¼ì´ë”©)
```
